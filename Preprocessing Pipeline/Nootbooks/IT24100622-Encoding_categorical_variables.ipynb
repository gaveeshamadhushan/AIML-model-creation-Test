```python
import pandas as pd
import numpy as np
import seaborn as sns  #visualisation
import matplotlib.pyplot as plt #visualisation
%matplotlib inline 
sns.set(color_codes=True)
```

```python
df=pd.read_csv("Time-Wasters on Social Media.csv")
```

```python
display(df.head())  #display first five rows
```

```python
df=df.rename(columns={"Sex":"Gender","Population":"Demographics","Social Media Platform":"Platform"})#Column renaming because column names are difficult to understand user.
df.head(5)
```

```python
df["Gender"]=df["Gender"].map({"Male":1,"Female":0,"Other":2}) #algorithms can understand and process the data correctly
```

```python
df.head(5)# after the changes display rows
```

```python
df["Demographics"]=df["Demographics"].map({"Rural":1,"Urban":0})
df.head(5)
```

```python
df["Platform"]=df["Platform"].map({"TikTok":1,"Instagram":2,"YouTube":3,"Facebook":4})
df.head(5)
```

```python
df.shape #Check the dimension of the DataFrame
```

```python
df.count #used to count number of rows
```

```python
df=df.dropna()#dropping the missing values
df.count #after the dropping count number of rows
```

```python
#boxplot for platforms
sns.boxplot(x=df['Platform'])
plt.show()
```

```python
#boxplot for addction levels
sns.boxplot(x=df['Addiction Level'])
plt.show()
```

```python
#Visualization - scatter plot
fig,ax = plt.subplots(figsize=(10,6))
ax.scatter(df['Addiction Level'],df['Age'])
ax.set_xlabel('Addiction Level')
ax.set_ylabel('Age')
plt.show()
```

```python
#Visualization - scatter plot
fig,ax = plt.subplots(figsize=(10,6))
ax.scatter(df['Platform'],df['Addiction Level'])
ax.set_xlabel('Platform')
ax.set_ylabel('Addiction Level')
plt.show()

#histogram for addction levels of platforms
df.Platform.value_counts().nlargest(40).plot(kind='bar',figsize=(10,5))
plt.title("Platform Vs Addiction Level")
plt.y_lable("Addiction Level")
plt.x_lable("Platform")
plt.show()
```
